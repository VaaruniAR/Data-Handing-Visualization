```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G=nx.DiGraph()
```

```python
G.add_nodes_from([1,2,3,4,5])
```

```python
G.add_edges_from([(1,2),(1,3),(2,3),(3,4),(5,1)])
```

```python
pos=nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_color='Skyblue',node_size=1400,edge_color='red',linewidths=1,font_size=15)
plt.show()
```
