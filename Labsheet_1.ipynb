```

import matplotlib.pyplot as plt


y = [30, 25, 20, 15]
labels = ["apples", "bananas", "cherries", "dates"]
myexplode = [0.2, 0, 0, 0.5]


plt.pie(y, labels=labels, explode=myexplode)
plt.show()
```

```
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=["r", "y", "b", "g", "b"]
sizes=[100, 200, 300, 40, 15]
plt.scatter(x, y, c=colors, s=sizes, alpha=0.8, marker="*", edgecolor="g")
plt.title("scatter", fontsize=20)
plt.xlabel("xaxis", fontsize=20)
plt.ylabel("yaxis", fontsize=20)
plt.show()
```
